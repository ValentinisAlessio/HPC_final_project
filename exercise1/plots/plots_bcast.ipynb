```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# BROADCAST ALGORITHM

## Binary tree algorithm

```python
# Read in the data
bin_tree_core = pd.read_csv('../results_def/bcast_bintree_results_wu_core.csv')
bin_tree_socket = pd.read_csv('../results_def/bcast_bintree_results_wu_socket.csv')
bin_tree_node = pd.read_csv('../results_def/bcast_bintree_results_wu_node.csv')

# Merge the data
bin_tree = pd.concat([bin_tree_core, bin_tree_socket, bin_tree_node])

bin_tree.head()
bin_tree.tail()
bin_tree_socket.tail()
```

```python
# plot the results concerning the 1 MessageSize
df = bin_tree[bin_tree['MessageSize'] == 1]

sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Allocation', data=df, marker='o')
plt.axvline(64, color='grey', linestyle='--')
plt.axvline(128, color='grey', linestyle='--')
plt.axvline(256, color='grey', linestyle='--')
plt.xlabel('Number of Processes')
plt.ylabel('Time in microSeconds')
plt.title('Time for 1 Byte Broadcast')

# Inctrease dimensions of the plot
fig = plt.gcf()
fig.set_size_inches(14, 8)

plt.show()

# save the plot
fig.savefig('bcast_bintree_1byte.png')
```
