```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data
df = pd.read_csv('../results/bcast_bintree_results_wu.csv')
df.head()
```

```python
# plot the results concerning the 1 MessageSize
df1 = df[df['MessageSize'] == 1]

sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Allocation', data=df1)
plt.xlabel('Number of Processes')
plt.ylabel('Time in microSeconds')
plt.title('Time for 1 Byte Broadcast')

plt.show()
```
