```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Barrier algorithm

## Default algorithm

```python
barrier_default_core= pd.read_csv('../results_def/barrier_default_results_wu_core.csv')
barrier_default_socket= pd.read_csv('../results_def/barrier_default_results_wu_socket.csv')
barrier_default_node= pd.read_csv('../results_def/barrier_default_results_wu_node.csv')

barrier_default = pd.concat([barrier_default_core, barrier_default_socket, barrier_default_node])
# shift the nessagesize column to the avg latency column
barrier_default['Avg Latency(us)'] = barrier_default['MessageSize']
# drop the old messagesize column
barrier_default = barrier_default.drop('MessageSize', axis=1)

barrier_default.head()
```

```python
# plot results

sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Allocation', data=barrier_default, marker='o')
plt.xlabel('Number of Processes')
plt.ylabel('Time in microSeconds')
plt.title('Time for Barrier with Default Algorithm')

# Inctrease dimensions of the plot
fig = plt.gcf()
fig.set_size_inches(14, 8)

plt.show()
```
