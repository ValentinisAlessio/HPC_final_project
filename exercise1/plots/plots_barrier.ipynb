```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Barrier algorithm

## Default algorithm

```python
barrier_default_core= pd.read_csv('../results_def/barrier_default_results_wu_core.csv')
barrier_default_socket= pd.read_csv('../results_def/barrier_default_results_wu_socket.csv')
barrier_default_node= pd.read_csv('../results_def/barrier_default_results_wu_node.csv')

barrier_default = pd.concat([barrier_default_core, barrier_default_socket, barrier_default_node])



barrier_default.head()
```

```python
# plot results

sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Allocation', data=barrier_default, marker='o')
plt.axvline(64, color='grey', linestyle='--')
plt.axvline(128, color='grey', linestyle='--')
plt.axvline(256, color='grey', linestyle='--')
plt.xlabel('Number of Processes')
plt.ylabel('Time in microSeconds')
plt.title('Time for Barrier with Default Algorithm')

# Inctrease dimensions of the plot
fig = plt.gcf()
fig.set_size_inches(6, 4)

plt.show()

# save plot
fig.savefig('../plots/barrier_default.png', dpi=100)
```

## Global by core behavior

```python
# Merge the dataframes
data = pd.concat([barrier_default, barrier_linear, barrier_ring, barrier_bruck])

# Select the allocation by core
data_core = data[data['Allocation'] == 'core']

#plot the results
sns.set_style("whitegrid")
sns.set_context("paper")
sns.set_palette("colorblind")
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Algorithm', data=data_core, marker='o')
plt.axvline(64, color='grey', linestyle='--')
plt.axvline(128, color='grey', linestyle='--')
plt.axvline(256, color='grey', linestyle='--')
plt.xlabel('Number of Processes')
plt.ylabel('Time in microSeconds')
plt.title('Time for Barrier with different Algorithms')

# Inctrease dimensions of the plot
fig = plt.gcf()
fig.set_size_inches(10, 6)

plt.show()

# save plot
fig.savefig('../plots/barrier_core.png', dpi=100)
```
